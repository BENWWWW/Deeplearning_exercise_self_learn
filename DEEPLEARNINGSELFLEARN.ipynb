```python
from transformers import pipeline

sentiment_analyzer = pipeline('sentiment-analysis')
```

```python
def chatbot_response(message):
    """
    Analyzes the sentiment of a message and returns a response about universities.

    Args:
        message: The input message string.

    Returns:
        A response string based on the sentiment and keywords in the message.
    """
    university_info = {
        "admission": "University admissions typically require an application, transcripts, and standardized test scores like the SAT or ACT.",
        "programs": "Universities offer a wide range of programs including arts, sciences, engineering, business, and more.",
        "tuition": "Tuition costs vary greatly depending on the university, location, and whether you are an in-state or out-of-state student.",
        "campus life": "Campus life at universities is vibrant with various clubs, sports, social events, and opportunities for personal growth."
    }

    result = sentiment_analyzer(message)
    sentiment_label = result[0]['label']

    if sentiment_label != 'NEGATIVE':
        message_lower = message.lower()
        for keyword, info in university_info.items():
            if keyword in message_lower:
                return info
        return "Ask me about admissions, programs, tuitions and campus life"
    else:
        return "negative comment :("
```

```python
import gradio as gr

iface = gr.Interface(fn=chatbot_response,
                     inputs=gr.Textbox(label="Enter your message"),
                     outputs=gr.Textbox(label="Chatbot Response"),
                     title="University Information Chatbot")
iface.launch()
```
